```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys


from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
```

```python
driver = webdriver.Chrome('./chromedriver.exe')

driver2 = webdriver.Chrome('./chromedriver.exe')
```

```python
sitefreguesias="https://pt.wikipedia.org/wiki/Lista_de_freguesias_de_Portugal"
sitemunicipios="https://pt.wikipedia.org/wiki/Lista_de_munic%C3%ADpios_de_Portugal"

```

```python
driver.get(sitefreguesias)
```

```python
elems = driver.find_elements_by_xpath("//div[@class='mw-parser-output']//h3")

entries=[]

id=0
header=("location_id","name","type","parent_id","link")
len(elems)
for el in elems:
    
    name= el.find_element_by_xpath(".//a").text
    link= el.find_element_by_xpath(".//a").get_attribute('href')
    
    
    
    entries.append((id,name,"C",'',link))
    print("C:",name)
    fregs = el.find_elements_by_xpath("./following-sibling::ul[1]/li/a")
    
    
    fregid=id
    for freg in fregs:
        
        fregname=freg.text
        freglink= freg.get_attribute('href')
        
        if fregname==name:
            continue
        
        print("F:",fregname)
        entries.append((id,fregname,'F',fregid,freglink))
        fregid=fregid+1
    
    
    id=fregid+1
```

```python
df = pd.DataFrame (entries, columns = header)
```

```python
df.to_excel("fregs.xlsx",
             sheet_name='Sheet_name_1')
```

```python
lats=[]
longs=[]

for row in df.iterrows():
    
    driver.get(row[1]['link'])
    try:
        geolinkel = WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.XPATH, "//a[contains(@href,'geohack')]")))
    
    except:
        
        lat = ""
        long =  ""    
        print(lat,long,row[1]['link'])
    
        lats.append(lat)
        longs.append(long)
    
        continue
        
    geolink=  geolinkel.get_attribute('href')
    
    driver.get(geolink)
    
    lat = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@class='latitude']"))).text
    long =  WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//*[@class='longitude']"))).text
    
    print(lat,long,row[1]['link'])
    
    lats.append(lat)
    longs.append(long)
    
```

```python
df['lat']=lats
df['long']=longs

df.to_excel("fregs1.xlsx",
             sheet_name='Sheet_name_1')
```
